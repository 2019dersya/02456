```python
import os
import pandas as pd
import json
 
```

```python
##Path for all json you have 
listing = os.listdir("/Users/GabrielFournier/Downloads/AllJson/")


#Since for each of us the categories are not the same we need to compare them
CategoriesDict=[]
NewCategroiresDict=[]



freshNewOne={
    "licenses":[],
    "info":{},
    "categories": [],
    "images": [],
    "annotations": [],
    
}
k=0
for jsFile in listing:
    if(jsFile!='.DS_Store'):
        NewCategroiresDict=[]
        #Path of the specific json annotation file 
        
        path="/Users/GabrielFournier/Downloads/AllJson/"+jsFile
        print(jsFile)
        
        #read the json file
        f = open(path,)
        data = json.load(f)
        

        #Take the first json of the folder as a reference to setup the categories
        if(k==0):
            freshNewOne["licenses"]=data["licenses"]
            freshNewOne["info"]=data["info"]
            freshNewOne["categories"]=data["categories"]
            freshNewOne["images"]=data["images"]
            freshNewOne["annotations"]=data["annotations"]
            
            #Just to make a comparaison we need the categories of the first json file
            for j in range(len(data["categories"])):
                elt={}
                elt["id"] = data["categories"][j]["id"]
                elt["name"] = data["categories"][j]["name"]
                elt["supercategory"] = data["categories"][j]["supercategory"]
                CategoriesDict.append(elt)
            images=data["images"]
            
        #the others json will go in this part
        
        if(k!=0):
            newImages=[]
            newAnnotations=[]
            
            #same proccess to compare the categories
            for j in range(len(data["categories"])):
                elt={}
                elt["id"] = data["categories"][j]["id"]
                #print(data["categories"][j]["name"])
                elt["name"] = data["categories"][j]["name"]
                elt["supercategory"] = data["categories"][j]["supercategory"]
                NewCategroiresDict.append(elt)
                
            #If the categories are the same juste need to concatenate the images + annotation
            #need to update the differents ids of images
            #print("CategoriesDict == NewCategroiresDict",CategoriesDict == NewCategroiresDict)
            if (CategoriesDict == NewCategroiresDict):
               
                for j in range(len(data["images"])):
                    #take the image 
                    newImage=data["images"][j]

                    #update the id
                    newId=data["images"][j]["id"]+len(freshNewOne["images"])
                    newImage["id"]=newId
                    
                    newImages.append(newImage)
                
                
                ##ADD ANNOTATIONS FIELD ON THE NEW JSON
                for j in range(len(data["annotations"])):
                    #take the annotation 
                    newAnnotation=data["annotations"][j]

                    #update id
                    newId=data["annotations"][j]["id"]+len(freshNewOne["annotations"])
                    newAnnotation["id"]=newId

                    #update Image_id
                    newImageId=data["annotations"][j]["image_id"]+len(freshNewOne["images"])
                    newAnnotation["image_id"]=newImageId

                    newAnnotations.append(newAnnotation)
                
                
                ##ADD all the things at the end
                freshNewOne["images"]=freshNewOne["images"]+newImages
                freshNewOne["annotations"]=freshNewOne["annotations"]+newAnnotations
                
            #if the categories are differents there are some permutation to do 
            else:
                #print("============")
                #permutation part
                normal=[]
                new=[]
                for j in range(len(CategoriesDict)):
                    new.append(NewCategroiresDict[j]["name"])
                    normal.append(CategoriesDict[j]["name"])
                    #print(NewCategroiresDict[j]["name"])

                permutation=[]
                for elt in new:
                    #print(elt)
                    permutation.append(normal.index(elt))
                    #print(normal.index(elt))
                


                
                #ADD IMAGES FIELD ON THE NEW JSON
                for j in range(len(data["images"])):
                    newImage=data["images"][j]
                    #print(data["images"][j]["id"]+len(freshNewOne["images"])-1)
                    
                    
                    newId=data["images"][j]["id"]+len(freshNewOne["images"])
                    newImage["id"]=newId
                    
                    newImages.append(newImage)
                
                #print(freshNewOne["images"])
                
                ##ADD ANNOTATIONS FIELD ON THE NEW JSON
                for j in range(len(data["annotations"])):
                    newAnnotation=data["annotations"][j]

                    newId=data["annotations"][j]["id"]+len(freshNewOne["annotations"])
                    newAnnotation["id"]=newId

                    newImageId=data["annotations"][j]["image_id"]+len(freshNewOne["images"])
                    newAnnotation["image_id"]=newImageId
                    
                    #need to update all the category id 
                    newCategory_id=permutation[data["annotations"][j]["category_id"]-1]+1
                    newAnnotation["category_id"]=newCategory_id

                    newAnnotations.append(newAnnotation)
                
                
                ##ADD all the things at the end
                freshNewOne["images"]=freshNewOne["images"]+newImages
                freshNewOne["annotations"]=freshNewOne["annotations"]+newAnnotations

                
        with open('/Users/GabrielFournier/Downloads/AllJson/newJson.json', 'w') as outfile:
            json.dump(freshNewOne, outfile)


        k+=1
```
