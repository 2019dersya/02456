```python
import os
import pandas as pd
import json
import math
```

```python
path = "/Users/GabrielFournier/Downloads/AllJson/newJson.json"
f = open(path,)
data=json.load(f)
```

```python
trainData={
    "licenses":data["licenses"],
    "info":data["info"],
    "categories": data["categories"],
    "images": [],
    "annotations": [],   
}

testData={
    "licenses":data["licenses"],
    "info":data["info"],
    "categories": data["categories"],
    "images": [],
    "annotations": [],   
}
```

```python
numberoftrain = math.floor(len(data["images"])*0.8)
numberofTest=len(data["images"])-numberoftrain
print(numberoftrain)

trainImage=[]
testImage=[]

for k in range(0,numberoftrain):
    trainImage.append(data["images"][k])
    #print(k)

newId=1
for j in range(numberoftrain,len(data["images"])):
    image=data["images"][j]
    image["id"]=newId
    testImage.append(image)
    newId+=1
```

```python
trainData["images"]=trainImage
testData["images"]=testImage
```

```python
trainAnnotations=[]
testAnnotations=[]

newId=1

for k in range(len(data["annotations"])):

    if(data["annotations"][k]["image_id"] < numberoftrain+1):
        trainAnnotations.append(data["annotations"][k])
        #print(data["annotations"][k]["image_id"])
    else:
        #print("gm")
        annotations=data["annotations"][k]
        annotations["id"]=newId
        annotations["image_id"]=annotations["image_id"]-numberoftrain
        testAnnotations.append(annotations)
        newId+=1
        
    
```
